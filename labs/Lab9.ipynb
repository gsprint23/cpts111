## Task 0
For this task, we are going to add functionality to our code from Lab5 Task0. Lab5 Task0 stated the following:

>Write a program to compute the BMI (body mass index) for a person given their weight (in pounds) and height (in inches). The BMI formula is as follows:
$$BMI = \frac{weight}{height^{2}} \times 703$$

>Define *at least* the following functions in your solution:
1. `get_user_weight()`: Prompts the user to enter their weight in pounds and returns the weight.
1. `get_user_height()`: Prompts the user to enter their height in *feet and inches*. Convert the feet and inches to total inches and return this value.
1. `compute_BMI()`: Accepts two arguments, `weight_pounds` and `height_inches` and returns the resulting BMI.
1. `display_BMI()`: Accepts three arguments, `weight_pounds`, `height_inches`, and `bmi` and displays the results.
1. `display_BMI_classification()`, that accepts a single argument, `bmi`, and displays a BMI classification based on the following table (from [online BMI calculator](http://www.bmi-calculator.net/)):

|BMI|Classification|
|-|-|
|18.5 or less|	Underweight|
|18.5 to 24.99|	Normal Weight|
|25 to 29.99|	Overweight|
|30 to 34.99|	Obesity (Class 1)|
|35 to 39.99|	Obesity (Class 2)|
|40 or greater|	Morbid Obesity|

>Check your computation is correct by comparing it to an [online BMI calculator](http://www.bmi-calculator.net/).

>Example output:

>```
Please enter the weight in pounds: 
150
Please enter the height in feet: 
5
Please enter the remaining height in inches: 
0
The BMI for a person with weight 150.00 pounds and height 60.00 inches is: 29.29
This BMI is in the category: Overweight
```

### Lab 5 Task 0 Code Solution

````python
def get_user_weight():
    '''
    
    '''
    print("Please enter the weight in pounds: ")
    pounds = float(input())
    return pounds
    
def get_user_height():
    '''
    
    '''
    print("Please enter the height in feet: ")
    feet = int(input())
    print("Please enter the remaining height in inches: ")
    inches = int(input())
    total_inches = feet * 12 + inches    
    return total_inches

def compute_BMI(weight_pounds, height_inches):
    '''
    
    '''
    bmi = (weight_pounds / (height_inches * height_inches)) * 703
    return bmi
    
def display_BMI(weight_pounds, height_inches, bmi):
    '''
    
    '''
    print("The BMI for a person with weight ", end="")
    print("%.2f pounds and height %.2f inches is: %.2f" %(weight_pounds, height_inches, bmi))
    
def display_BMI_message(bmi):
    '''

    '''
    print("This BMI is in the category: ", end="")
    if bmi <= 18.5:
        print("Underweight")
    elif bmi < 25:
        print("Normal Weight")
    elif bmi < 30:
        print("Overweight")
    elif bmi < 35:
        print("Obesity (Class 1)")
    elif bmi < 40:
        print("Obesity (Class 2)")
    else:
        print("Morbid Obesity")

def main():
    '''
    
    '''
    weight = get_user_weight()
    height = get_user_height()
    bmi = compute_BMI(weight, height)
    display_BMI(weight, height, bmi)
    display_BMI_message(bmi)

main()
````
