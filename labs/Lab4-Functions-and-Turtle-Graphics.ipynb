## Task 0
For this task (and other tasks in this lab), we are going to use the [Turtle graphics](https://docs.python.org/3.4/library/turtle.html) module. Imagine a turtle standing at the origin (0,0) of an x-y canvas. For fun, let's name this turtle Crush (from Finding Nemo):
<img src="https://raw.githubusercontent.com/gsprint23/cpts111/master/labs/figures/crush_turtle.png" width="500"/>
Crush is carrying a pen in his mouth. When we give Crush commands like `forward(15)` or `right(15)`, we can move Crush around, thus drawing a picture. Take a moment and read through the list of available turtle commands [here](https://docs.python.org/3.4/library/turtle.html#overview-of-available-turtle-and-screen-methods).

Here are some common turtle methods (methods are functions that are associated with an object):
1. `forward(<number of pixels)`: moves the turtle forward a specified number of pixels
1. `backward(<number of pixels)`: moves the turtle backward a specified number of pixels
1. `left(<number of degrees>)`: rotates the turtle counter clockwise a specified number of degrees
1. `right(<number of degrees>)`: rotates the turtle clockwise a specified number of degrees
1. `setpos(<x coordinate>, <y coordinate>)`: moves the turtle to the x, y coordinate arguments
1. `speed(<integer speed>)`: increase the animation speed. An argument of 0 indicates the fastest drawing speed. See other speed options[here](https://docs.python.org/3.4/library/turtle.html#turtle.speed)
1. `penup()`: lifts the pen up so the turtle does not draw on the canvas when he/she moves
1. `pendown()`: lowers the pen to the canvas so the turtle does draw when he/she moves
1. `pencolor(<color string>)`: changes the pen color. See the available pen colors [here](https://www.tcl.tk/man/tcl8.4/TkCmd/colors.htm)
1. `pensize(<integer width>)`: changes the pen size to the specified width
1. `shape(<turtle string>)`: change the turtle "cursor". See the available shapes [here](https://docs.python.org/3.4/library/turtle.html#turtle.shape)
1. Much more!

Here is example turtle code that moves Crush to coordinate (5, 20) and then draws a thick green line 200 pixels (steps) in length:

````python
# import the turtle graphics module
import turtle

# change the pen to the shape of a turtle
turtle.shape("turtle")

# dont draw a line while we move the turtle to the first coordinate
turtle.penup()

# move the turtle
turtle.setpos(5, 20)

# change pen color and size
turtle.pencolor("green")
turtle.pensize(10)

# prepare to start drawing
turtle.pendown()

# move the turtle
turtle.forward(200)

# if running Python as a script: start the GUI (graphical user interface) event loop
turtle.mainloop()
````

### Background Info
The `turtle` module contains a function called [`onkey(<function>, <key string>)`](https://docs.python.org/3.5/library/turtle.html#turtle.onkey) that *binds* a function to a keystroke. For example:

```
import turtle
turtle.onkey(forward, "Up")
```

This line of code binds a function named `forward()` to the "Up" key on your keyboard. This means that if the user presses the "Up" key, the function `forward()` will be called behind the scenes by Python.

Lastly, we need to let `turtle` graphics know to "listen" for user input events by calling a function called [`listen()`](https://docs.python.org/3.5/library/turtle.html#turtle.listen):
```
turtle.listen()
turtle.mainloop()
```
